# LSTM

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import tensorflow.keras.optimizers as optimizers
```

```python
data = pd.read_csv('parking_data_cleaned.csv')
df = pd.DataFrame(data)
df.head()
```

```python
df['garagecode'].unique()
```

```python
# Define train and test size
train_size = 0.8

# Get unique garage codes
garage_codes = df['garagecode'].unique().tolist()

# Initialize a dictionary to hold training and testing data for each garage
garage_data = {}

# Iterate over unique garage codes
for garage in garage_codes:
    garage_data[garage] = {'train': {}, 'test': {}}
    
    # Filter DataFrame for the current garage
    garage_df = df[df['garagecode'] == garage]
    
    # Get unique days of the week for the current garage
    week_days = garage_df['weekday'].unique().tolist()
    
    # Iterate over days of the week
    for day in week_days:
        # Filter DataFrame for the current day of the week
        day_df = garage_df[garage_df['weekday'] == day]
        
        # Determine the index to split the data for training and testing
        split_index = int(len(day_df) * train_size)
        
        # Split the DataFrame into train and test DataFrames
        train_df, test_df = day_df.iloc[:split_index], day_df.iloc[split_index:]
        
        # Reset index for both train and test DataFrames
        train_df.reset_index(drop=True, inplace=True)
        test_df.reset_index(drop=True, inplace=True)
        
        # Store train and test DataFrames in the dictionary
        garage_data[garage]['train'][day] = train_df
        garage_data[garage]['test'][day] = test_df
```

```python
train_dfs
```

```python
# create target cols list
target_cols = ['capacity']

# create feature cols list
#feat_cols = df.columns
feat_cols = ['capacity','weekday_cos','weekday_sin']
#feat_cols = feat_cols.drop(['vehiclecount', 'totalspaces', 'garagecode', 'update_dt', 'date', 'time', 'weekday', 'day', 'month', 'year'])
#feat_cols = feat_cols.to_list()

seq_length = 96 #24 hours with a 30min timestep
ph = 48
```

```python
input_seqs = []
output_seqs = []

for df in train_dfs:
    for i in range(len(df) - seq_length - ph):
        sequence = df[feat_cols].iloc[i:i+seq_length]
        target = df[target_cols].iloc[i+seq_length+ph-1]

        input_seqs.append(sequence)
        output_seqs.append(target)

input_seqs = np.array(input_seqs, dtype=object).astype(np.float32)
output_seqs = np.array(output_seqs, dtype=object).astype(np.float32).reshape(-1)

print(f'input data shape: {input_seqs.shape}')
print(f'target shape: {output_seqs.shape}')
```

```python
# test data sequencing
val_seqs = []
val_outputs = []

for df in test_dfs:
    for i in range(seq_length, len(df) - ph):
        seq = df[feat_cols].iloc[i-seq_length: i]
        target = df[target_cols].iloc[i+ph-1]

        val_seqs.append(seq)
        val_outputs.append(target)

val_seqs = np.array(val_seqs, dtype=object).astype(np.float32)
val_outputs = np.array(val_outputs, dtype=object).astype(np.float32).reshape(-1)
#val_outputs = te_outputs.reshape(-1)


print(f'input data shape: {val_seqs.shape}')
print(f'target shape: {val_outputs.shape}')
```

```python
seq_length = input_seqs.shape[1]  # Example: 3 if you're using 3 time steps
nb_features = input_seqs.shape[2]  # Number of features in your input
nb_out = 1  # For regression tasks, typically 1 output node

model = Sequential()

# Add first LSTM layer
model.add(LSTM(input_shape=(seq_length, nb_features),
               units=5, activation='relu',return_sequences=True))
model.add(Dropout(0.2)) 

# Add second LSTM layer
model.add(LSTM(units=3,activation='relu', return_sequences=False))
model.add(Dropout(0.2))

# Add output layer
model.add(Dense(units=nb_out, activation='relu'))

# Compile the model
optimizer = optimizers.Adam(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer=optimizer, metrics=['mse'])

# Model summary
print(model.summary())

# Define callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=0, mode='min'),
    ModelCheckpoint('model.h5', monitor='val_loss', save_best_only=True, mode='min', verbose=0)
]

# Fit the network
history = model.fit(input_seqs, output_seqs,epochs=100, batch_size=50, validation_split=0.1, verbose=2, callbacks=callbacks)
# List all data in history
print(history.history.keys())
```

```python
plt.figure(figsize=(10, 5))

rmse = np.sqrt(history.history['mse'])
val_rmse = np.sqrt(history.history['val_mse'])

# Plot RMSE
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot(rmse)
plt.plot(val_rmse)
plt.title('Model RMSE')
plt.ylabel('RMSE')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Plot Loss
plt.subplot(1, 2, 2)  # (rows, columns, panel number)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

# Show the figure
plt.show()
```

# Validation

```python
scores_test = model.evaluate(val_seqs, val_outputs, verbose=2)
print('\nMSE: {}\n'.format(scores_test))

y_pred_test = model.predict(val_seqs)
y_true_test = val_outputs
```
